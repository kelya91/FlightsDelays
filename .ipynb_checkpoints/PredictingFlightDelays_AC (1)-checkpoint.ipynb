```python
import pandas as pd
import numpy as np
from pprint import pprint
```

```python
flights_df = pd.read_csv("flights.csv")
airlines_df = pd.read_csv("airlines.csv")
airports_df = pd.read_csv("airports.csv")
```

```python
combined_df = pd.merge(flights_df, airlines_df, left_on='AIRLINE', right_on='IATA_CODE')
combined_df = pd.merge(combined_df, airports_df, left_on='ORIGIN_AIRPORT', right_on='IATA_CODE')
```

```python
combined_df = combined_df[['YEAR', 'MONTH', 'DAY', 'DAY_OF_WEEK', 'AIRLINE_x', 'FLIGHT_NUMBER', 'ORIGIN_AIRPORT', 'LATITUDE', 'LONGITUDE', 'CITY', 'STATE', 'COUNTRY', 'DESTINATION_AIRPORT', 'SCHEDULED_DEPARTURE', 'DEPARTURE_TIME', 'SCHEDULED_TIME', 'SCHEDULED_ARRIVAL', 'DISTANCE', 'ARRIVAL_DELAY', 'DEPARTURE_DELAY']]
correlated_df = combined_df[['YEAR', 'MONTH', 'DAY', 'DAY_OF_WEEK', 'AIRLINE_x', 'FLIGHT_NUMBER', 'ORIGIN_AIRPORT', 'LATITUDE', 'LONGITUDE', 'CITY', 'STATE', 'COUNTRY', 'DESTINATION_AIRPORT', 'SCHEDULED_DEPARTURE', 'DEPARTURE_TIME', 'SCHEDULED_TIME', 'SCHEDULED_ARRIVAL', 'DISTANCE', 'ARRIVAL_DELAY']]
```

```python
corr_matrix = correlated_df.corr().abs()
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))
to_drop = [column for column in upper.columns if any(upper[column] > 0.80)]
```

```python
to_drop
```

```python
combined_df = combined_df.drop('DEPARTURE_TIME', axis=1)
combined_df = combined_df.drop('DISTANCE', axis=1)
combined_df = combined_df.drop('YEAR', axis=1)
combined_df.head()
```

```python
combined_df['DEPARTURE_DELAY'] = np.where(combined_df['DEPARTURE_DELAY'] > 0, 1, 0).astype('int')
```

```python
combined_df.head(10)
```

```python
combined_df = combined_df.dropna()
```

```python
'''Reduce dataset to 25% due to memory errors'''

combined_df = combined_df.sample(frac=0.025, replace=True, random_state=1)
```

```python
combined_df = pd.get_dummies(combined_df)
```

```python
combined_df.head()
```

```python
X = combined_df.drop('DEPARTURE_DELAY', axis=1)
y = combined_df['DEPARTURE_DELAY']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, stratify=y)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model
```

```python
model.fit(X_train, y_train)
```

```python
print(f"Training Data Score: {model.score(X_train, y_train)}")
print(f"Testing Data Score: {model.score(X_test, y_test)}")
```

```python
predictions = model.predict(X_test)
overallresults = pd.DataFrame({"Prediction": predictions, "Actual": y_test})
overallresults
```

```python
'''Based on model, there is a 65% chance of predicting an accurate result for a delayed flight'''

delayedresults = overallresults[overallresults['Actual'] == 1]
delayedbalance = delayedresults['Prediction'].sum() / len(delayedresults)
delayedbalance
```

```python
'''Save model'''

import pickle

filename = 'predictingflightdelays.sav'
pickle.dump(model, open(filename, 'wb'))
loaded_model = pickle.load(open(filename, 'rb'))
```

```python
result = loaded_model.score(X_test, y_test)
print(result)
```
